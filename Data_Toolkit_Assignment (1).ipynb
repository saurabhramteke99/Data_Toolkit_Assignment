Que_2. How does broadcasting work in NumPy ?

Ans: Broadcasting is NumPy’s way of automatically expanding arrays with smaller dimensions so that they can be used with arrays of larger dimensions during arithmetic operations (like addition, subtraction, multiplication, etc.).

Instead of manually reshaping or copying data, NumPy "broadcasts" the smaller array across the larger one.

Broadcasting Rules:


1)When operating on two arrays, NumPy compares their shapes element-wise starting from the rightmost dimension, and works if:

2)The dimensions are equal, or One of the dimensions is 1

3)If the shapes don’t match and can’t be broadcast according to these rules, you'll get a ValueError.

Example:

```
A = np.array([[1], [2], [3]])     # Shape: (3, 1)
B = np.array([[10, 20, 30, 40]])  # Shape: (1, 4)

C = A + B
# Shape: (3, 4)
Result:
[[11 21 31 41]
 [12 22 32 42]
 [13 23 33 43]]

```

Que_6. What are the differences between NumPy arrays and Python lists ?

Ans: Here’s a descriptive breakdown of the key differences between NumPy arrays and Python lists:

1) Data Type Consistency :

NumPy Array:
a) All elements are of the same data type (e.g., all integers, all floats).

b) This uniformity allows for faster computation and efficient memory usage.

 Python List:
a) Can hold multiple data types (e.g., int, float, string all in the same list).

b) More flexible, but this makes operations slower and less optimized.

2) Performance and Speed
NumPy Array:

a) Much faster for numerical operations.

b) Built on C, and uses vectorized operations (operations done at C speed behind the scenes).

Python List:

a) Slower when performing mathematical operations.

b) Requires explicit loops for most computations.

3) Memory Efficiency

NumPy Array:

a)Stored in a contiguous block of memory.

b)Takes less memory than lists, especially for large datasets.

Python List:

a) Each element is a separate Python object.

b) Uses more memory due to overhead of storing mixed data types and pointers.

4) Functionality and Operations

NumPy Array:
a) Supports element-wise operations, matrix multiplication, broadcasting, advanced indexing, and multi-dimensional slicing.

b) Ideal for scientific computing, data analysis, and machine learning.

Python List:
a) Limited to basic operations like append, pop, slicing, and looping.

b) Does not support mathematical or linear algebra operations directly.

5) Dimensionality

NumPy Array:

a) Can be multi-dimensional (1D, 2D, 3D, etc.).

b) Essential for matrix operations, image data, tensors, etc.

Python List:
Can be nested (list of lists), but lacks built-in tools to handle multi-dimensional data efficiently.

Que_21. Why is NumPy’s array slicing faster than Python’s list slicing ?

Ans: NumPy’s array slicing is faster and more efficient than Python’s list slicing due to several key reasons:

*  NumPy Arrays Are Stored in Contiguous Memory :

1)NumPy arrays use homogeneous data types (all elements are of the same type) and are stored in a single block of memory.

2) This allows direct access to any element using pointer arithmetic, which is extremely fast.

3)  Python lists, on the other hand, are heterogeneous and store references to objects in memory scattered all over the place. Accessing elements means following pointers to potentially different memory locations.

* Slicing in NumPy Creates Views, Not Copies:

1)When you slice a NumPy array, it returns a view of the original data (i.e., no new array is created).

2)This is super efficient because it just gives you a new window into the existing array without copying any data.

3) Python list slicing, in contrast, creates a new list and copies the sliced elements, which takes more time and memory.

* NumPy Is Written in C (Under the Hood):

1) NumPy is implemented in C, which is much closer to the hardware level and significantly faster than Python’s interpreted code.

2) Slicing and other operations are heavily optimized and vectorized in compiled code.

3) Python lists use higher-level, interpreted operations, which are slower.

* Low-Level Optimizations :

1) NumPy uses vectorized operations and internal strides to determine how to jump from one element to the next without looping through each one.

2) These stride tricks make slicing and other operations lightning-fast.

3) Python lists don’t have stride-based optimization — slicing is just copying elements one-by-one.
